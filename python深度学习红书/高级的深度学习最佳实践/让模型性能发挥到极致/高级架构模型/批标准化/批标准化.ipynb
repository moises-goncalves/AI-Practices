## 批标准化
# 在训练过程中均值和方差会变化 Keras.BatchNormalization层会跟踪这些变化 然后在训练中动态调整
# 有助于梯度传播 特别是那些特别深的网络

# BatchNormalization层通常在卷积层或者全链接层之后使用
# 接受一个参数axis 指定要归一化的轴
# 一般对最后一个轴进行归一化 即axis=-1 如果有人的数据格式是channels_first 则axis=1

## 当batch_size比较小的时候 可以使用批再标准化 效果会比批标准化好一些

```python
import keras.layers

conv_model = keras.models.Sequential()
conv_model.add(keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
conv_model.add(keras.layers.BatchNormalization()) # 在卷积层之后使用

conv_model.add(keras.Dense(32, activation='relu'))
conv_model.add(keras.layers.BatchNormalization())  # 在全链接层之后使用
```
