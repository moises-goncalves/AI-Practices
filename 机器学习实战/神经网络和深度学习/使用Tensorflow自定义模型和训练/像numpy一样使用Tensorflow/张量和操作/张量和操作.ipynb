# 12.2.1 张量和操作

## 使用tf.constant()创建张量

```python
import tensorflow as tf
matrix1 = tf.constant([[1, 2, 3], [4, 5, 6]])
print(matrix1)
```

```python
matrix1.shape
```

```python
matrix1.dtype
```

```python
matrix1[:, 1:]
```

```python
print(matrix1 + 10)
```

```python
# 开平方
print(tf.square(matrix1))
```

```python
# 计算矩阵与自身转置的乘法 @是python里面的矩阵乘法运算符
print(matrix1 @ tf.transpose(matrix1))
```
