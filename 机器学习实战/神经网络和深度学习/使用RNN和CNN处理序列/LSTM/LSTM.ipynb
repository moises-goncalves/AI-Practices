```python
# 推荐的 Keras 定义方式，以消除警告
from tensorflow import keras

model = keras.models.Sequential([
    # 显式 Input 层作为模型的第一层
    keras.Input(shape=[None, 1]),
    keras.layers.SimpleRNN(50, return_sequences=True),
    keras.layers.SimpleRNN(50, return_sequences=True),
    keras.layers.SimpleRNN(20),
    keras.layers.Dense(1)
])
```
