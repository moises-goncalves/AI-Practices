```python
import keras
class LNSimpleRNNCell(keras.layers):
    def __init__(self, units, activation='tanh', **kwargs):
        super().__init__(**kwargs)
        self.units = units
        self.activation = keras.activations.get(activation)
        self.kernel_initializer = keras.initializers.get('glorot_uniform')
        self.output_size = units
        self.simple_rnn_cell = keras.layers.SimpleRNNCell(units, activation=activation)

    def call(self, inputs, states):
        outputs, new_states = self.simple_rnn_cell(inputs, states)
        norm_outputs = self.activation(self.layer_norm(outputs))
        return norm_outputs, [norm_outputs]

model = keras.models.Sequential([
    keras.layers.RNN(LNSimpleRNNCell(32), input_shape=(None, 10), return_sequences=True),
    keras.layers.RNN(LNSimpleRNNCell(32), return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(10))
])
```
