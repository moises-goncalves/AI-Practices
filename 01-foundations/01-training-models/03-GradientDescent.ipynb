## ğŸ“– ç¬¬1éƒ¨åˆ†ï¼šç†è®ºèƒŒæ™¯

### ä»€ä¹ˆæ˜¯æ¢¯åº¦ä¸‹é™ï¼Ÿ

**æ¢¯åº¦ä¸‹é™**æ˜¯ä¸€ç§è¿­ä»£ä¼˜åŒ–ç®—æ³•ï¼Œç”¨äºæ‰¾åˆ°ä½¿æŸå¤±å‡½æ•°æœ€å°åŒ–çš„å‚æ•°å€¼ã€‚å®ƒé€šè¿‡æ²¿ç€æ¢¯åº¦çš„åæ–¹å‘æ›´æ–°å‚æ•°æ¥é€æ­¥é€¼è¿‘æœ€ä¼˜è§£ã€‚

### æ ¸å¿ƒæ€æƒ³

æƒ³è±¡ä½ ç«™åœ¨å±±ä¸Šï¼Œæƒ³è¦åˆ°è¾¾å±±è°·æœ€ä½ç‚¹ã€‚æœ€å¿«çš„æ–¹æ³•æ˜¯å§‹ç»ˆæœç€æœ€é™¡å³­çš„ä¸‹å¡æ–¹å‘èµ°ã€‚

### æ•°å­¦å…¬å¼

**å‚æ•°æ›´æ–°è§„åˆ™**ï¼š
$$\theta := \theta - \eta \nabla_\theta J(\theta)$$

å…¶ä¸­ï¼š
- $\theta$ æ˜¯æ¨¡å‹å‚æ•°
- $\eta$ æ˜¯å­¦ä¹ ç‡ (learning rate)
- $\nabla_\theta J(\theta)$ æ˜¯æŸå¤±å‡½æ•°çš„æ¢¯åº¦

### çº¿æ€§å›å½’çš„æ¢¯åº¦

å¯¹äºMSEæŸå¤±å‡½æ•°ï¼š
$$J(\theta) = \frac{1}{2m}\sum_{i=1}^{m}(h_\theta(x^{(i)}) - y^{(i)})^2$$

æ¢¯åº¦ä¸ºï¼š
$$\nabla_\theta J(\theta) = \frac{1}{m}X^T(X\theta - y)$$

> ğŸ’¡ **å…³é”®æ´å¯Ÿ**: æ¢¯åº¦æŒ‡å‘å‡½æ•°å¢é•¿æœ€å¿«çš„æ–¹å‘ï¼Œæ‰€ä»¥æˆ‘ä»¬æ²¿ç€æ¢¯åº¦çš„**åæ–¹å‘**ç§»åŠ¨ä»¥æœ€å°åŒ–æŸå¤±ã€‚

### æ­¥éª¤1: å¯¼å…¥åº“å’Œé…ç½®

```python
# ============================================================
# å¯¼å…¥å¿…è¦çš„åº“
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ============================================================
# é…ç½®
# ============================================================

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

print("âœ“ é…ç½®å®Œæˆ")
```

### æ­¥éª¤2: ç”Ÿæˆæ•°æ®

```python
# ============================================================
# ç”Ÿæˆçº¿æ€§æ•°æ®
# ============================================================

# æ•°æ®å‚æ•°
m = 100  # æ ·æœ¬æ•°
TRUE_THETA_0 = 4.0  # çœŸå®æˆªè·
TRUE_THETA_1 = 3.0  # çœŸå®æ–œç‡

# ç”Ÿæˆæ•°æ®: y = 4 + 3x + å™ªå£°
X = 2 * np.random.rand(m, 1)
y = TRUE_THETA_0 + TRUE_THETA_1 * X + np.random.randn(m, 1) * 0.5

# æ·»åŠ åç½®é¡¹ x0=1
X_b = np.c_[np.ones((m, 1)), X]

print(f"âœ“ ç”Ÿæˆ {m} ä¸ªæ ·æœ¬")
print(f"  çœŸå®å‚æ•°: Î¸â‚€={TRUE_THETA_0}, Î¸â‚={TRUE_THETA_1}")
print(f"  X_bå½¢çŠ¶: {X_b.shape}")
```

```python
# ============================================================
# å¯è§†åŒ–æ•°æ®
# ============================================================

plt.figure(figsize=(10, 5))

plt.scatter(X, y, alpha=0.6, s=50, edgecolors='k', linewidth=0.5)
plt.plot([0, 2], [TRUE_THETA_0, TRUE_THETA_0 + 2*TRUE_THETA_1], 
         'g--', linewidth=2, label=f'çœŸå®çº¿: y = {TRUE_THETA_0} + {TRUE_THETA_1}x')
plt.xlabel('X')
plt.ylabel('y')
plt.title('ç”Ÿæˆçš„çº¿æ€§æ•°æ®', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### æ­¥éª¤3: å®ç°æ‰¹é‡æ¢¯åº¦ä¸‹é™

```python
# ============================================================
# æ‰¹é‡æ¢¯åº¦ä¸‹é™å®ç°
# ============================================================

def batch_gradient_descent(X_b, y, eta=0.1, n_iterations=1000, 
                           record_history=False):
    """
    æ‰¹é‡æ¢¯åº¦ä¸‹é™ç®—æ³•
    
    å‚æ•°:
        X_b: å¸¦åç½®é¡¹çš„ç‰¹å¾çŸ©é˜µ (m, n+1)
        y: ç›®æ ‡å‘é‡ (m, 1)
        eta: å­¦ä¹ ç‡
        n_iterations: è¿­ä»£æ¬¡æ•°
        record_history: æ˜¯å¦è®°å½•å†å²
    
    è¿”å›:
        theta: æœ€ç»ˆå‚æ•°
        history: å†å²è®°å½•ï¼ˆå¦‚æœrecord_history=Trueï¼‰
    """
    m = len(y)
    
    # éšæœºåˆå§‹åŒ–å‚æ•°
    theta = np.random.randn(2, 1)
    
    history = {'theta': [theta.copy()], 'cost': []}
    
    for iteration in range(n_iterations):
        # 1. è®¡ç®—é¢„æµ‹å€¼
        predictions = X_b.dot(theta)
        
        # 2. è®¡ç®—è¯¯å·®
        errors = predictions - y
        
        # 3. è®¡ç®—æ¢¯åº¦
        # æ¢¯åº¦ = (1/m) * X^T * (XÎ¸ - y)
        gradients = (2/m) * X_b.T.dot(errors)
        
        # 4. æ›´æ–°å‚æ•°
        theta = theta - eta * gradients
        
        if record_history:
            # è®°å½•å‚æ•°å’ŒæŸå¤±
            history['theta'].append(theta.copy())
            cost = (1/(2*m)) * np.sum(errors**2)
            history['cost'].append(cost)
    
    if record_history:
        return theta, history
    return theta

print("âœ“ æ‰¹é‡æ¢¯åº¦ä¸‹é™å‡½æ•°å®šä¹‰å®Œæˆ")
```

```python
# ============================================================
# è¿è¡Œæ¢¯åº¦ä¸‹é™
# ============================================================

# å‚æ•°è®¾ç½®
LEARNING_RATE = 0.1
N_ITERATIONS = 1000

# è¿è¡Œæ¢¯åº¦ä¸‹é™
theta_gd, history = batch_gradient_descent(
    X_b, y, 
    eta=LEARNING_RATE, 
    n_iterations=N_ITERATIONS,
    record_history=True
)

print("æ¢¯åº¦ä¸‹é™ç»“æœ:")
print(f"  ä¼°è®¡çš„Î¸â‚€: {theta_gd[0][0]:.6f} (çœŸå®å€¼: {TRUE_THETA_0})")
print(f"  ä¼°è®¡çš„Î¸â‚: {theta_gd[1][0]:.6f} (çœŸå®å€¼: {TRUE_THETA_1})")

# å¯¹æ¯”sklearnçš„ç»“æœ
lin_reg = LinearRegression()
lin_reg.fit(X, y)
print(f"\nSklearnç»“æœ:")
print(f"  Î¸â‚€: {lin_reg.intercept_[0]:.6f}")
print(f"  Î¸â‚: {lin_reg.coef_[0][0]:.6f}")
```

## ğŸ“Š ç¬¬3éƒ¨åˆ†ï¼šå¯è§†åŒ–æ¢¯åº¦ä¸‹é™è¿‡ç¨‹

```python
# ============================================================
# å¯è§†åŒ–æŸå¤±ä¸‹é™è¿‡ç¨‹
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# å·¦å›¾ï¼šæŸå¤±å‡½æ•°ä¸‹é™
axes[0].plot(history['cost'], 'b-', linewidth=2)
axes[0].set_xlabel('è¿­ä»£æ¬¡æ•°')
axes[0].set_ylabel('æŸå¤± (MSE)')
axes[0].set_title('æŸå¤±å‡½æ•°æ”¶æ•›è¿‡ç¨‹', fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].set_xlim(0, min(100, len(history['cost'])))  # åªæ˜¾ç¤ºå‰100æ¬¡

# å³å›¾ï¼šå‚æ•°æ”¶æ•›
theta_history = np.array(history['theta']).squeeze()
axes[1].plot(theta_history[:, 0], 'r-', linewidth=2, label='Î¸â‚€')
axes[1].plot(theta_history[:, 1], 'b-', linewidth=2, label='Î¸â‚')
axes[1].axhline(y=TRUE_THETA_0, color='r', linestyle='--', alpha=0.5, label=f'çœŸå®Î¸â‚€={TRUE_THETA_0}')
axes[1].axhline(y=TRUE_THETA_1, color='b', linestyle='--', alpha=0.5, label=f'çœŸå®Î¸â‚={TRUE_THETA_1}')
axes[1].set_xlabel('è¿­ä»£æ¬¡æ•°')
axes[1].set_ylabel('å‚æ•°å€¼')
axes[1].set_title('å‚æ•°æ”¶æ•›è¿‡ç¨‹', fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_xlim(0, min(100, len(theta_history)))

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# å¯è§†åŒ–å‚æ•°ç©ºé—´ä¸­çš„æ¢¯åº¦ä¸‹é™è·¯å¾„
# ============================================================

# åˆ›å»ºå‚æ•°ç½‘æ ¼
theta0_range = np.linspace(2, 6, 100)
theta1_range = np.linspace(1, 5, 100)
theta0_grid, theta1_grid = np.meshgrid(theta0_range, theta1_range)

# è®¡ç®—æ¯ä¸ªç‚¹çš„æŸå¤±å€¼
cost_grid = np.zeros_like(theta0_grid)
for i in range(len(theta0_range)):
    for j in range(len(theta1_range)):
        theta_temp = np.array([[theta0_range[i]], [theta1_range[j]]])
        cost_grid[j, i] = (1/(2*m)) * np.sum((X_b.dot(theta_temp) - y)**2)

plt.figure(figsize=(10, 8))

# ç»˜åˆ¶ç­‰é«˜çº¿
plt.contour(theta0_grid, theta1_grid, cost_grid, levels=30, cmap='viridis')
plt.colorbar(label='æŸå¤±å€¼')

# ç»˜åˆ¶æ¢¯åº¦ä¸‹é™è·¯å¾„
theta_hist = np.array(history['theta']).squeeze()
plt.plot(theta_hist[:50, 0], theta_hist[:50, 1], 'r.-', linewidth=1, markersize=5, label='æ¢¯åº¦ä¸‹é™è·¯å¾„')
plt.scatter(theta_hist[0, 0], theta_hist[0, 1], c='green', s=100, marker='o', zorder=5, label='èµ·ç‚¹')
plt.scatter(theta_gd[0], theta_gd[1], c='red', s=100, marker='*', zorder=5, label='ç»ˆç‚¹')
plt.scatter(TRUE_THETA_0, TRUE_THETA_1, c='blue', s=100, marker='x', zorder=5, label='çœŸå®å€¼')

plt.xlabel('Î¸â‚€')
plt.ylabel('Î¸â‚')
plt.title('å‚æ•°ç©ºé—´ä¸­çš„æ¢¯åº¦ä¸‹é™è·¯å¾„', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# ============================================================
# ä¸åŒå­¦ä¹ ç‡çš„å¯¹æ¯”
# ============================================================

learning_rates = [0.002, 0.02, 0.1, 0.5]

fig, axes = plt.subplots(2, 4, figsize=(16, 8))

for i, eta in enumerate(learning_rates):
    np.random.seed(42)  # ç¡®ä¿ç›¸åŒçš„åˆå§‹åŒ–
    
    theta, hist = batch_gradient_descent(
        X_b, y, eta=eta, n_iterations=50, record_history=True
    )
    
    # ä¸Šæ’ï¼šæŸå¤±æ›²çº¿
    axes[0, i].plot(hist['cost'], 'b-', linewidth=2)
    axes[0, i].set_xlabel('è¿­ä»£æ¬¡æ•°')
    axes[0, i].set_ylabel('æŸå¤±')
    axes[0, i].set_title(f'Î· = {eta}', fontweight='bold')
    axes[0, i].grid(True, alpha=0.3)
    
    # ä¸‹æ’ï¼šæ‹Ÿåˆçº¿å˜åŒ–
    theta_hist = np.array(hist['theta'])
    axes[1, i].scatter(X, y, alpha=0.3, s=30)
    
    # ç»˜åˆ¶å‡ æ¡è¿­ä»£è¿‡ç¨‹ä¸­çš„çº¿
    colors = plt.cm.viridis(np.linspace(0, 1, 10))
    for j, idx in enumerate([0, 5, 10, 20, -1]):
        if idx < len(theta_hist):
            t = theta_hist[idx]
            axes[1, i].plot([0, 2], [t[0], t[0] + 2*t[1]], 
                           color=colors[j*2], alpha=0.7, 
                           label=f'iter={idx}' if idx != -1 else f'iter={len(theta_hist)-1}')
    
    axes[1, i].set_xlim(-0.1, 2.1)
    axes[1, i].set_xlabel('X')
    axes[1, i].set_ylabel('y')
    axes[1, i].legend(fontsize=7)
    axes[1, i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nå­¦ä¹ ç‡å½±å“:")
print("  - Î·å¤ªå°: æ”¶æ•›å¾ˆæ…¢ï¼Œéœ€è¦å¾ˆå¤šè¿­ä»£")
print("  - Î·é€‚ä¸­: å¿«é€Ÿç¨³å®šæ”¶æ•›")
print("  - Î·å¤ªå¤§: å¯èƒ½éœ‡è¡ç”šè‡³å‘æ•£")
```
