```python
from tensorflow import keras
from functools import partial

RegurizedDense = partial(keras.layers.Dense, kernel_regularizer=keras.regularizers.l2(0.001),
                         activation = "elu",kernel_initializer="he_normal")

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.BatchNormalization(),
    RegurizedDense(256),
    RegurizedDense(128),
    RegurizedDense(10, activation="softmax",kernel_initializer="glorot_uniform")
])
```
